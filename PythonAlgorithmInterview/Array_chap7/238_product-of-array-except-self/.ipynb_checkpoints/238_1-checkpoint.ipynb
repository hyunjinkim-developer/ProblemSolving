```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        output = []
        p = 1
        # multiply from left to right
        for i in range(0, len(nums)):
            output.append(p)
            p = p * nums[i] # * location of this line
            # the condition of the problem was product of all the elements of nums 'except' nums[i]
            # current value should not be included in multiplication
            # push the value to rightward for 1 index

        p = 1
        # mutiply the value of mutiplied from right to left
        for i in range(len(nums) - 1, 0 - 1, -1):
            output[i] = output[i] * p
            p = p * nums[i] # * location of this line
            # current value should not be included in multiplication
            # push the value to leftward for 1 index
        return output
```
