# Sieve of Eratosthenes:
## How to find all prime numbers with in a certain range in the shortest period of time
<a href=https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes>Reference</a>


```python
from typing import List

def prime_list(n: int) -> List[int]:
    sieve = [True] * n

    # biggest divisor of n is smaller than sqrt(n),
    # ***** check until sqrt(n) *****
    m = int(n ** 0.5) # sqrt(n)
    for i in range(2, m + 1): # *important
        if sieve[i] == True:
            # multiple of prime number is not a prime number
            for j in range(i+i, n, i): # *important
                sieve[j] = False

    # all prime numbers 'under' n : [2, n)
    return [i for i in range(2, n) if sieve[i] == True] # *important

display(prime_list(100))
```
