```python
def print_matrix(matrix):
    for r in range(len(matrix)):
        print(matrix[r])

def solution(friends, gifts):
    answer = 0

    LEN = len(friends)
    friend_d = {k: v for k, v in zip(friends, list(range(LEN)))}
    this_month = [[0 for _ in range(LEN)]
                    for _ in range(LEN)]
    gift_idx = [0 for _ in range(LEN)]
    next_month = [0 for _ in range(LEN)]
    for gift in gifts:
        from_, to_ = gift.split()
        from_idx, to_idx = friend_d[from_], friend_d[to_]
        this_month[from_idx][to_idx] += 1

    # Get gift index
    for r in range(LEN):
        give = sum(this_month[r])
        gift_idx[r] += give
    for c in range(LEN):
        get = 0
        for r in range(LEN):
            get += this_month[r][c]
        gift_idx[c] -= get

    for r in range(LEN):
        for c in range(r + 1, LEN):
            A, B = r, c
            if this_month[A][B] > this_month[B][A]:
                next_month[A] += 1
            elif this_month[A][B] < this_month[B][A]:
                next_month[B] += 1
            else:
                A_gift_idx = gift_idx[A]
                B_gift_idx = gift_idx[B]
                if A_gift_idx > B_gift_idx:
                    next_month[A] += 1
                elif A_gift_idx < B_gift_idx:
                    next_month[B] += 1
                else:
                    continue
    answer = max(next_month)
    return answer
```

```python
from itertools import combinations, product
from collections import Counter, defaultdict
from functools import reduce
from operator import mul

def calculate_prob(dice_list):
    probability = defaultdict(int)

    list_ = list()
    for dice in dice_list:
        each_dice = list()
        for k, v in Counter(dice).items():
            each_dice.append((k, v))
        list_.append(each_dice)
    for element in list(product(*list_)):
        picked_sum = sum(list(zip(*element))[0])
        count_for_each_picekd = list(zip(*element))[1]
        picked_count = reduce(mul, count_for_each_picekd)
        probability[picked_sum] += picked_count
    return probability

def solution(dice):
    answer = []

    COUNT = len(dice)
    # dice_idxes = set([for _ in range(COUNT)])
    DICE_COUNT = COUNT // 2
    if COUNT >= 2:
        pairs = list(combinations(list(range(COUNT)), DICE_COUNT))

    max_win = 0
    for pair in pairs:
        a_dice = list()
        b_dice = list()
        win, tie, lose = 0, 0, 0
        for i in range(COUNT):
            if i in pair:
                a_dice.append(dice[i])
            else:
                b_dice.append(dice[i])

        a_prob = calculate_prob(a_dice)
        b_prob = calculate_prob(b_dice)
        for a in a_prob.keys():
            for b in b_prob.keys():
                count = a_prob[a] * b_prob[b]
                if a > b:
                    win += count
                elif a < b:
                    lose += count
                else:
                    tie += count

        if win > max_win:
            max_win = win
            convert_idx = list()
            for i in list(pair):
                convert_idx.append(i + 1)
            answer = convert_idx
    return answer
```
