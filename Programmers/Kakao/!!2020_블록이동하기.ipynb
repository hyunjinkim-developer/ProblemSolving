# Find shortest path using BFS
- ! idea : robot takes 2 space and can rotate
- Add (robot location and time) in queue
- Check visited location with set() data structure
    - "visited" contains locations that are already visited or already queued
- Add (robot location and time in "visited" right after it is queued
- *tuple : to unpack tuple into elements

<a href=https://velog.io/@tjdud0123/블록-이동하기-2020-카카오-공채-python>Reference</a>

```python
from collections import deque

board = list()
LEN = 0

def in_range(r, c):
    global LEN
    return 0 <= r < LEN and 0 <= c < LEN


def find_movable_positions(pos1, pos2):
    global board

    movable_pos = list()

    p1_r, p1_c = pos1
    p2_r, p2_c = pos2

    # # from up in clockwise
    dirs = [(-1, 0), (1, 0), (0, 1), (0, -1)]
    for dr, dc in dirs:
        p1_nr, p1_nc = p1_r + dr, p1_c + dc
        p2_nr, p2_nc = p2_r + dr, p2_c + dc
        if (not in_range(p1_nr, p1_nc)) or (not in_range(p2_nr, p2_nc)): continue
        if (board[p1_nr][p1_nc] == 1) or (board[p2_nr][p2_nc] == 1): continue
        movable_pos.append(((p1_nr, p1_nc), (p2_nr, p2_nc)))

    # Rotation
    # Robot in horizontal
    if p1_r == p2_r:
        up, down = -1, 1
        for dr in [up, down]:
            p1_nr = p1_r + dr
            p2_nr = p2_r + dr

            if (not in_range(p1_nr, p1_c)) or (not in_range(p2_nr, p2_c)): continue
            if (board[p1_nr][p1_c] == 0) and (board[p2_nr][p2_c] == 0):
                movable_pos.append((pos1, (p1_nr, p1_c))) # axis pos1
                movable_pos.append((pos2, (p2_nr, p2_c))) # axis pos2
    # Robot in vertical
    if p1_c == p2_c:
        left, right = -1, 1
        for dc in [left, right]:
            p1_nc = p1_c + dc
            p2_nc = p2_c + dc
            if (not in_range(p1_r, p1_nc)) or (not in_range(p2_r, p2_nc)): continue
            if (board[p1_r][p1_nc] == 0) and (board[p2_r][p2_nc] == 0):
                movable_pos.append(((p1_r, p1_nc), pos1)) # axis pos1
                movable_pos.append(((p2_r, p2_nc), pos2)) # axis pos2
    return movable_pos


def bfs():
    global board, LEN

    time = 0
    start = ((0, 0), (0, 1))
    end = (LEN - 1, LEN - 1)

    q = deque([(*start, time)]) # * := unpack tuple
    visited = set([start])

    while q:
        pos1, pos2, time = q.popleft()
        if pos1 == end or pos2 == end:
            break
        for next_pos in find_movable_positions(pos1, pos2):
            if next_pos in visited: continue # Already visited or added in queue
            q.append((*next_pos, time + 1)) # * := unpack tuple
            visited.add(next_pos) # ! Reduce time complexity: check whether it is in queue

    return time


def solution(b):
    answer = 0

    """
    지도는 가장 왼쪽, 상단의 좌표를 (1, 1)로 하며
    지도 내에 표시된 숫자 "0"은 빈칸을 "1"은 벽을 나타냅니다.
    로봇은 처음에 좌표 (1, 1) 위치에서 가로방향으로 놓여있는 상태로 시작
    회전하는 방향(축이 되는 칸으로부터 대각선 방향에 있는 칸)에는 벽이 없어야 합니다
    로봇이 한 칸 이동하거나 90도 회전하는 데는 걸리는 시간은 정확히 1초
    """
    global board, LEN
    board, LEN = b, len(b)

    answer = bfs()
    return answer
```
