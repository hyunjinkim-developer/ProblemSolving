## Lock을 가운데 두고 Padding을 사용해서 영역을 확장(Board)
- Key를 Board 좌측상단부터 확인
    - attach() : Key가 Board에 적용된 상태
    - lock_key_matched() : Lock영역에 Key가 꽂혀 있는지 (Lock + Key == 1) 확인
    - detach() : Board에서 Key적용을 해제한 상태
- 이런 방식을 적용하면 총 4번의 roatate_clockwise()에 대해서만 고려하면 됨

```python
Key, Lock, Board = list(), list(), list()
M, N, B = 0, 0, 0

# For debugging purposes
def print_matrix(matrix):
    for r in range(len(matrix)):
        for c in range(len(matrix[0])):
            print(matrix[r][c], end="\t")
        print()
    print("-"*10)


def in_range(r, c, len_):
    return 0 <= r < len_ and 0 <= c < len_


def roatate_clockwise():
    global Key, M

    result = [[0 for _ in range(M)] for _ in range(M)]
    for r in range(M):
        for c in range(M):
            result[c][M - 1 - r] = Key[r][c]
    Key = result


def lock_key_matched():
    global Board, M, N
    for r in range(M, M + N):
        for c in range(M, M + N):
            if Board[r][c] != 1:
                return False
    return True

def attach(start_r, start_c):
    global Board, B, Key, M

    if not in_range(start_r + M, start_c + M, B): return False

    for r in range(M):
        for c in range(M):
            nr, nc = start_r + r, start_c + c
            Board[nr][nc] = Board[nr][nc] + Key[r][c]
    return True

def detach(start_r, start_c):
    global Board, B, Key, M

    for r in range(M):
        for c in range(M):
            nr, nc = start_r + r, start_c + c
            Board[nr][nc] = Board[nr][nc] - Key[r][c]

def move_lock():
    global Board, B

    for r in range(B):
        for c in range(B):
            if attach(r, c):
                if lock_key_matched() == True:
                    return True
                detach(r, c)
    return False


def solution(key, lock):
    answer = False

    # 열쇠는 회전과 이동이 가능
    # 자물쇠 영역을 벗어난 부분에 있는 열쇠의 홈과 돌기는 자물쇠를 여는 데 영향을 주지 않지만,
    # 자물쇠 영역 내에서는 열쇠의 돌기 부분과 자물쇠의 홈 부분이 정확히 일치해야
    # M(key)은 항상 N(lock) 이하
    global Key, M, Lock, N, Board, B
    Key, Lock = key, lock
    M, N = len(key), len(lock)

    Board = [[0 for c in range(2 * M + N)]
             for r in range(2 * M + N)]
    B = 2 * M + N
    for r in range(M, M + N):
        for c in range(M, M + N):
            Board[r][c] = Lock[r - M][c - M]

    for i in range(4):
        if move_lock() == True:
            return True
        roatate_clockwise()

    return False
```
