```python
# Solution 1:
from collections import deque

def replace_dotdot(new_id):
    new_id = deque(new_id)
    result = deque()
    queue = list()
    while new_id:
        c = new_id.popleft()
        if c == ".":
            queue.append(c)
        else:
            if len(queue) != 0:
                result.append(".")
                queue = list()
                result.append(c)
            else:
                result.append(c)
    if len(queue) != 0:
        result.append(".")
    new_id = "".join(result)
    return new_id

def solution(new_id):
    answer = ''

    # new_id의 모든 대문자를 대응되는 소문자로 치환합니다.
    new_id = new_id.lower()
    # new_id에서 알파벳 소문자, 숫자, 빼기(-), 밑줄(_), 마침표(.)를 제외한 모든 문자를 제거합니다.
    for c in "~!@#$%^&*()=+[{]}:?,<>/":
        new_id = new_id.replace(c, "")
    # new_id에서 마침표(.)가 2번 이상 연속된 부분을 하나의 마침표(.)로 치환합니다.
    new_id = replace_dotdot(new_id)
    # new_id에서 마침표(.)가 처음이나 끝에 위치한다면 제거합니다.
    new_id = new_id.rstrip(".")
    new_id = new_id.lstrip(".")
    # new_id가 빈 문자열이라면, new_id에 "a"를 대입합니다.
    if len(new_id) == 0:
        new_id += "a"
    # new_id의 길이가 16자 이상이면, new_id의 첫 15개의 문자를 제외한 나머지 문자들을 모두 제거합니다.
    if len(new_id) >= 16:
        new_id = new_id[:15]
      # 만약 제거 후 마침표(.)가 new_id의 끝에 위치한다면 끝에 위치한 마침표(.) 문자를 제거합니다.
        new_id = new_id.rstrip(".")
    # new_id의 길이가 2자 이하라면, new_id의 마지막 문자를 new_id의 길이가 3이 될 때까지 반복해서 끝에 붙입니다.
    if len(new_id) <= 2:
        while len(new_id) < 3:
            new_id += new_id[-1]
    answer = new_id
    return answer
```

```python
# Solution 2:
def solution(new_id):
    answer = ''

    # 1 new_id의 모든 대문자를 대응되는 소문자로 치환합니다.
    new_id = new_id.lower()

    # 2 알파벳 소문자, 숫자, 빼기(-), 밑줄(_), 마침표(.)를 제외한 모든 문자를 제거합니다.
    for c in new_id:
        if c.isalpha() or c.isdigit() or c in ["-", "_", "."]: # !idea
            answer += c

    # 3 마침표(.)가 2번 이상 연속된 부분을 하나의 마침표(.)로 치환합니다.
    while ".." in answer: # !idea
        answer = answer.replace("..", ".")

    # 4 new_id에서 마침표(.)가 처음이나 끝에 위치한다면 제거합니다.
    if answer[0] == ".":
        answer = answer[1:] if len(answer) > 1 else "." # for index out of range
    if answer[-1] == ".":
        answer = answer[:-1]

    # 5 빈 문자열이라면, new_id에 "a"를 대입합니다.
    if answer == "":
        answer += "a"

    # 6
    # 길이가 16자 이상이면, new_id의 첫 15개의 문자를 제외한 나머지 문자들을 모두 제거합니다.
    # 만약 제거 후 마침표(.)가 new_id의 끝에 위치한다면 끝에 위치한 마침표(.) 문자를 제거합니다.
    if len(answer) >= 16:
        answer = answer[:15]
        if answer[-1] == ".":
            answer = answer[:-1]

    # 7 길이가 2자 이하라면, new_id의 마지막 문자를 new_id의 길이가 3이 될 때까지 반복해서 끝에 붙입니다.
    while len(answer) < 3:
        answer += answer[-1]

    return answer
```
