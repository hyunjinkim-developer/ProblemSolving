```python
"""
On progress
"""
from collections import defaultdict, deque

def to_minute(string):
    h, m = tuple(map(int, string.split(":")))
    return 60 * h + m

def minute_to_string(minute):
    h, m = divmod(minute, 60)
    # string.ljust(length, character)
    # := left align the string
    # specified character (space is default) as the fill character.
    h, m = str(h).rjust(2, "0"), str(m).rjust(2, "0")
    print(":".join([h, m]))
    return ":".join([h, m])


def solution(n, t, m, timetable):
    answer = ''
    """
    셔틀 운행 횟수 n,
    셔틀 운행 간격 t,
    한 셔틀에 탈 수 있는 최대 크루 수 m
    """

    # Sort all crew's arrived time
    timetable = list(to_minute(crew) for crew in timetable)
    timetable.sort()
    timetable = deque(timetable)

    commute = defaultdict(list)
    # 셔틀은 09:00부터 총 n회 t분 간격으로 역에 도착
    shuttle_starts = to_minute("09:00")
    current_shuttle = shuttle_starts
    total_availables = n * m

    shuttle_no = 0
    idx = 0
    while timetable:
        crew_arrived_at = timetable.popleft()

        if crew_arrived_at <= current_shuttle:
            # Current shuttle is full
            if len(commute[current_shuttle]) == m:
                # See if the crew can get the next shuttle
                shuttle_no += 1
                if n <= shuttle_no:
                    break
                current_shuttle = current_shuttle + t
            commute[current_shuttle].append(crew_arrived_at)
        # if crew arrived later than current shuttle
        else:
            # See if the crew can get the next shuttle
            shuttle_no += 1
            if n <= shuttle_no:
                break
            current_shuttle = current_shuttle + t
            commute[current_shuttle].append(crew_arrived_at)
        idx += 1

        minute_to_string(current_shuttle)
        print(idx, end=" ")
        minute_to_string(crew_arrived_at)
    print(commute)



    # # 무사히 셔틀을 타고 사무실로 갈 수 있는 제일 늦은 도착 시각을 출력
    # If seat for Con is not available
    if len(commute[current_shuttle]) == m:
        answer = minute_to_string(crew_arrived_at - 1)
    else:
        answer = minute_to_string(current_shuttle)
    return answer
```
