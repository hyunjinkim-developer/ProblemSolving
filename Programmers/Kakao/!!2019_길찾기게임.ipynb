# Create binary tree
# Tree traversal in pre-order, post-order
    - pre-order := root -> left -> right
    - post-order := left -> right -> root
    - idea for traversal:
        1. add root in pre-order
        2. Among nodes y next biggest than root_y
            2-1. traverse x smaller than root_x
            2-2. traverse x bigger than root_x
            2-3. ! STOP after right node is traversed
        3. add root in post-order

```python
"""
Create tree with deque() (for performance to pop root from the tree)
reference: https://www.hamadevelop.me/algorithm-kakao2019/
"""

from collections import deque
import sys
sys.setrecursionlimit(10**5)
"""
In the question maximum depth of tree is 1000
but, functions are stacked in the runtime.
"""

levels = list()
pre_order = list()
post_order = list()


def traversal(tree, levels_idx):
    global levels, pre_order, post_order

    root = tree.popleft()
    root_index, root_info = root
    root_x, root_y = root_info[0], root_info[1]

    # pre_order := root -> left -> right
    pre_order.append(root_index)

    if tree: # if tree is not empty
        for i in range(len(tree)):
            index, info = tree[i]
            x, y = info[0], info[1]
            if y == levels[levels_idx + 1]:
                if x < root_x:
                    traversal(deque([x for x in tree if x[1][0] < root_x]), levels_idx + 1)
                else:
                    traversal(deque([x for x in tree if x[1][0] > root_x]), levels_idx + 1)
                    break # left -> right, no more recursive traversal

    # post_order := left -> right -> root
    post_order.append(root_index)


def solution(nodeinfo):
    global levels

    # Create Binary Tree
    # Get node level(y) and sort in DESC
    levels = sorted({info[1] for info in nodeinfo}, reverse=True)
    # Put index(start from 1) on each node
    nodes = list(zip(range(1, len(nodeinfo) + 1), nodeinfo))
    # Sort node in y DESC, if the same y x in ASC
    def custom_sort(item):
        index, info = item
        x, y = info[0], info[1]
        return y * -1, x
    nodes = sorted(nodes, key=custom_sort)
    tree = deque(nodes)

    traversal(tree, 0)
    answer = [pre_order, post_order]
    return answer
```

```python
"""
Solution 2:
Create tree with class
Reference: https://blog.naver.com/cjsencks/221774157760
"""
import sys
sys.setrecursionlimit(10**5)

class Node:
    def __init__(self, x, idx):
        self.x = x
        self.idx = idx
        self.left = None
        self.right = None

class Tree:
    def __init__(self):
        self.head = None
        self.preorder_result = list()
        self.postorder_result = list()

    def insert(self, x, idx):
        current_node = self.head
        if self.head is None:
            self.head = Node(x, idx)
        else:
            while True:
                if x < current_node.x:
                    if current_node.left: # 이미 left node가 존재
                        current_node = current_node.left # current_node를 root로 그 아래부터 다시 탐색
                    else:
                        current_node.left = Node(x, idx) # left에 새 노드 추가하고 loop break
                        break
                else:
                    if current_node.right: # 이미 right node가 존재
                        current_node = current_node.right # current_node를 root로 두고 그 아래부터 다시 탐색
                    else:
                        current_node.right = Node(x, idx) # right에 새 노드 추가하고 loop break
                        break

    def pre_order(self):
        def _pre_order(root):
            if root is None:
                pass
            else:
                # root -> left -> right
                self.preorder_result.append(root.idx)
                _pre_order(root.left)
                _pre_order(root.right)
        # Start from the root of the whole tree
        _pre_order(self.head)

    def post_order(self):
        def _post_order(root):
            if root is None:
                pass
            else:
                # left -> right -> root
                _post_order(root.left)
                _post_order(root.right)
                self.postorder_result.append(root.idx)
        # Start from the root of the whole tree
        _post_order(self.head)


def solution(nodeinfo):
    answer = [[]]

    # Create index
    nodeinfo = list(zip(list(range(1, len(nodeinfo) + 1)), nodeinfo))
    # Sort y in DESC
    def custom_sort(item):
        idx, info = item
        x, y = info[0], info[1]
        return y * -1
    nodeinfo = sorted(nodeinfo, key=custom_sort)

    # Create tree
    T = Tree()
    for i in range(len(nodeinfo)):
        T.insert(nodeinfo[i][1][0], nodeinfo[i][0]) # x, idx
    # Pre_order traversal
    T.pre_order()
    # Post_order traversal
    T.post_order()

    answer = [T.preorder_result, T.postorder_result]
    return answer
```
