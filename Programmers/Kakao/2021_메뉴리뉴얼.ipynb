```python
"""
Solution1:
Find combinations with all menus
"""

from collections import defaultdict, deque
from itertools import combinations
global Orders

def count_orders(availables):
    global Orders

    count_d = defaultdict(int)
    for menu_set in availables:
        for order in Orders:
            if set(menu_set).issubset(set(order)):
                count_d[menu_set] += 1
    sorted_count_d = sorted(count_d.items(), key=lambda item: item[1] * -1)

    course_candidates = list()
    if len(sorted_count_d) > 0:
        _, max_count = sorted_count_d[0]

        # 최소 2명 이상의 손님으로부터 주문된 단품메뉴 조합에 대해서만 코스요리 메뉴 후보에 포함
        if max_count >= 2:
            sorted_count_d = deque(sorted_count_d)
            while sorted_count_d:
                menu_set, count = sorted_count_d.popleft()
                if count < max_count:
                    break
                course_candidates.append(menu_set)
    return course_candidates


def solution(orders, course):
    answer = []
    global Orders
    Orders = orders

    all_menu = set()
    for order in Orders:
        all_menu |= set(order)

    availables = list()
    result = list()
    for course_len in course:
        availables = list(combinations(all_menu, course_len))
        result.extend(count_orders(availables))

    # 정답은 각 코스요리 메뉴의 구성을 문자열 형식으로 배열에 담아
    # 사전 순으로 오름차순 정렬해서 return 해주세요.
    for menu_set in result:
        answer.append("".join(sorted(menu_set)))
    answer = sorted(answer)
    return answer
```

```python
"""
Solution 2:
Find combinations with each order and union possible set
which can reduce time complexity
"""

from collections import defaultdict, deque
from itertools import combinations
global Orders

def count_orders(availables):
    global Orders

    count_d = defaultdict(int)
    for menu_set in availables:
        for order in Orders:
            if set(menu_set).issubset(set(order)):
                count_d[menu_set] += 1
    sorted_count_d = sorted(count_d.items(), key=lambda item: item[1] * -1)

    course_candidates = list()
    if len(sorted_count_d) > 0:
        _, max_count = sorted_count_d[0]

        # 최소 2명 이상의 손님으로부터 주문된 단품메뉴 조합에 대해서만 코스요리 메뉴 후보에 포함
        if max_count >= 2:
            sorted_count_d = deque(sorted_count_d)
            while sorted_count_d:
                menu_set, count = sorted_count_d.popleft()
                if count < max_count:
                    break
                course_candidates.append(menu_set)
    return course_candidates


def solution(orders, course):
    answer = []
    global Orders
    Orders = orders

    result = list()
    for course_len in course:
        availables = set()
        for order in Orders:
            # ! set() is not ordered, so sorted(set()) can remove redundancies
            availables |= set(combinations(sorted(set(order)), course_len))
        result.extend(count_orders(availables))

    # 정답은 각 코스요리 메뉴의 구성을 문자열 형식으로 배열에 담아
    # 사전 순으로 오름차순 정렬해서 return 해주세요.
    for menu_set in result:
        answer.append("".join(sorted(menu_set)))
    answer = sorted(answer)
    return answer
```
