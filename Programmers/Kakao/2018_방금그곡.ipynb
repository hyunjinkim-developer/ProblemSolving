```python
"""
Solution 1:
Use each note(C, C#, D, D#, E, E#, F, F#, G, G#, A, A#, B) as it is
To find match compare two lists(played_cords, m_cords)
"""
def convert_to_cords(notes):
    cords = list()
    for note in notes:
        if note == "#":
            last_cord = cords.pop()
            cords.append(f"{last_cord}#")
        else:
            cords.append(note)
    return cords


def find_match(played_cords, m):
    m_cords = convert_to_cords(m)

    played_cords_length = len(played_cords)
    m_cords_length = len(m_cords)

    if played_cords_length < m_cords_length:
        return False
    elif played_cords_length == m_cords_length:
        return played_cords == m_cords

    for p_idx in range(played_cords_length - m_cords_length + 1):
        count = 0
        for m_idx in range(m_cords_length):
            if m_cords[m_idx] != played_cords[p_idx + m_idx]:
                break
            else:
                count += 1
        if count == m_cords_length:
            return True
    return False


def solution(m, musicinfos):
    answer = ''

    played_music = dict()
    for idx, music in enumerate(musicinfos):
        start_time, end_time, title, notes = tuple(music.split(","))

        cords = convert_to_cords(notes)

        s_h, s_m = tuple(map(int, start_time.split(":")))
        e_h, e_m = tuple(map(int, end_time.split(":")))
        total_time = 60 * (e_h - s_h) + (e_m - s_m)

        played_cords = list()
        time = total_time
        if time > len(cords):
            while time:
                if time < len(cords):
                    played_cords.extend(cords[:time])
                    break
                else:
                    played_cords.extend(cords)
                    time -= len(cords)
        else:
            played_cords = cords[:time]

        played_music[title] = (played_cords, total_time, idx)

    # 조건이 일치하는 음악이 여러 개일 때에는 라디오에서 재생된 시간이 제일 긴 음악 제목을 반환한다.
    # 재생된 시간도 같을 경우 먼저 입력된 음악 제목을 반환한다.
    # 조건이 일치하는 음악이 없을 때에는 “(None)”을 반환한다.
    matched = dict()
    for title, (played_cords, played_time, idx) in played_music.items():
        if find_match(played_cords, m):
            matched[title] = (played_time, idx)

    if len(matched) == 0:
        answer = '(None)'
    else:
        def custom_sort(item):
            key, value = item
            played_time, idx = value
            return played_time * -1, idx
        sorted_matched = sorted(matched.items(), key=custom_sort)
        answer = sorted_matched[0][0]
    return answer
```

```python
"""
Solution 2:
Convert note with # into lowercase
Find match cords with m_string in played_cords_string
"""
def convert_to_cords(notes):
    cord_d = {"C#": "c", "D#": "d", "F#": "f", "E#": "e", "G#": "g", "A#": "a"}

    cords = list()
    for note in notes:
        if note == "#":
            last_cord = cords.pop()
            original_cord = f"{last_cord}#"
            cords.append(cord_d[original_cord])
        else:
            cords.append(note)
    return cords


def solution(m, musicinfos):
    answer = ''

    played_music = dict()
    for idx, music in enumerate(musicinfos):
        start_time, end_time, title, notes = tuple(music.split(","))

        cords = convert_to_cords(notes)

        s_h, s_m = tuple(map(int, start_time.split(":")))
        e_h, e_m = tuple(map(int, end_time.split(":")))
        total_time = 60 * (e_h - s_h) + (e_m - s_m)

        played_cords = list()
        time = total_time
        if time > len(cords):
            while time:
                if time < len(cords):
                    played_cords.extend(cords[:time])
                    break
                else:
                    played_cords.extend(cords)
                    time -= len(cords)
        else:
            played_cords = cords[:time]
        # Convert played_cords to string for efficient comparision with m
        played_cords = "".join(played_cords)
        played_music[title] = (played_cords, total_time, idx)

    # 조건이 일치하는 음악이 여러 개일 때에는 라디오에서 재생된 시간이 제일 긴 음악 제목을 반환한다.
    # 재생된 시간도 같을 경우 먼저 입력된 음악 제목을 반환한다.
    # 조건이 일치하는 음악이 없을 때에는 “(None)”을 반환한다.
    matched = dict()
    m = "".join(convert_to_cords(m))
    for title, (played_cords, played_time, idx) in played_music.items():
        if m in played_cords:
            matched[title] = (played_time, idx)

    if len(matched) == 0:
        answer = '(None)'
    else:
        def custom_sort(item):
            key, value = item
            played_time, idx = value
            return played_time * -1, idx
        sorted_matched = sorted(matched.items(), key=custom_sort)
        answer = sorted_matched[0][0]
    return answer
```
